# На шести вершинах

```python
graphs_dict6 = get_graphs_dict(6, max_edges=True)
signatures = get_all_signatures(graphs_dict6)
```

Cигнатур, как мы видим, довольно много, выведем некоторые из них: те, у которых сигнатура имеет длину 6, наибольшая степень 4 и таких вершин не более одной (чтоб избежать заведомо связных графов)

```python
def sgn6_to_print(sgn):
    if len(sgn) < 6 or max(sgn) == 5:
        return False
    if len([x for x in sgn if x == 4]) < 2:
        return True
```

```python
for signature in signatures[31:]:
    if sgn6_to_print(signature):
        plot_graphs_by_signature(signature, graphs_dict=graphs_dict6)
    # plt.show()
```
